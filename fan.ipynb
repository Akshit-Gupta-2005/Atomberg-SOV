```python
from serpapi import GoogleSearch
from youtube_comment_downloader import YoutubeCommentDownloader
import yt_dlp


def scrape_youtube_videos(query="smart fan", max_results=5):
    query += " india" 
    params = {
        "engine": "youtube",
        "search_query": query,
        "api_key": SERPAPI_KEY,
        "gl": "in",
        "hl": "en",
    }

    search = GoogleSearch(params)
    results = search.get_dict()

    videos = []
    for video in results.get("video_results", []):
        videos.append({
            "title": video.get("title"),
            "link": video.get("link"),
            "views": video.get("views"),
            "published": video.get("published"),
            "channel": video.get("channel", {}).get("name")
        })
        if len(videos) >= max_results:
            break

    return videos

def get_video_description(video_url):
    ydl_opts = {
        'quiet': True,
        'skip_download': True,
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(video_url, download=False)
            return info.get("description", "")
    except Exception as e:
        print(f"Error fetching description: {e}")
        return ""

def fetch_comments(video_url, max_comments=1):
    downloader = YoutubeCommentDownloader()
    comments = []
    try:
        for comment in downloader.get_comments_from_url(video_url):
            comments.append(comment['text'])
            if len(comments) >= max_comments:
                break
    except Exception as e:
        print(f"Error fetching comments: {e}")
    return comments

def analyze_brand_mentions(videos, brands):
    mention_counts = {brand: 0 for brand in brands}

    for video in videos:
        title = video['title']
        channel = video['channel']
        link = video['link']

        description = get_video_description(link)
        comments = fetch_comments(link)

        combined_text = f"{title} {channel} {description} {' '.join(comments)}".lower()

        print(f"\nVideo: {title}")
        print(f" Channel: {channel}")
        print(f" Link: {link}")
        print(f" Description Snippet: {description[:200]}")
        print(f" Top Comments:")
        for c in comments:
            print(f"   - {c}")
        print(" Checking mentions...")

        for brand in brands:
            if brand.lower() in combined_text:
                print(f"Found mention of '{brand}'")
                mention_counts[brand] += 1

        print("=" * 50)

    return mention_counts

if __name__ == "__main__":
    print("Scraping YouTube for query: 'smart fan'")
    videos = scrape_youtube_videos(query="smart fan", max_results=5)

    print(f"\nAnalyzing {len(videos)} videos for brand mentions...")
    mentions = analyze_brand_mentions(videos, brands)

    print("\nBrand mentions in YouTube videos for query: 'smart fan'")
    for brand in brands:
        print(f"{brand}: {mentions[brand]} mentions")
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

def get_sentiment_score(text):
    if not text:
        return 0
    return sia.polarity_scores(text)['compound'] 

def classify_sentiment(score, threshold=0.05):
    if score >= threshold:
        return "positive"
    elif score <= -threshold:
        return "negative"
    return "neutral"
```

```python
def compute_weight(platform, meta):
    if platform == "youtube":
        views = meta.get("views", 0)
        comments = meta.get("comments_count", 0)
        return (views/1000) + (comments/10) 
    elif platform == "google":
        rank = meta.get("rank", 10)
        return max(1, (11 - rank)) / 10 
    elif platform == "amazon":
        reviews = meta.get("reviews_count", 0)
        rating = meta.get("rating", 0)
        return (rating/5) * (1 + reviews/100)
    return 1
```

```python
def process_platform_data(results, platform, brands):
    
    processed = []

    for idx, item in enumerate(results, start=1):
        text_parts = []
        meta = {}

        if platform == "google":
            text_parts.append(item.get("title", ""))
            text_parts.append(item.get("snippet", ""))
            meta["rank"] = idx 
            weight = max(1, (11 - idx)) / 10  

        elif platform == "youtube":
            text_parts.append(item.get("title", ""))
            text_parts.append(item.get("channel", ""))
            text_parts.append(item.get("description", "")) 
            text_parts.extend(item.get("comments", []))
            meta["views"] = int(item.get("views", 0))
            meta["comments_count"] = len(item.get("comments", []))
            weight = (meta["views"]/1000) + (meta["comments_count"]/10)

        elif platform == "amazon":
            text_parts.append(item.get("title", ""))
            meta["rating"] = item.get("rating", 0)
            meta["reviews_count"] = item.get("reviews_count", 0)
            weight = (meta["rating"]/5) * (1 + meta["reviews_count"]/100)

        combined_text = " ".join(text_parts).lower()

        for brand in brands:
            if brand.lower() in combined_text:
                score = get_sentiment_score(combined_text)
                sentiment = classify_sentiment(score)
                processed.append({
                    "brand": brand.lower(),
                    "platform": platform,
                    "weight": weight,
                    "sentiment": sentiment
                })
                break  

    return processed
```

```python
keywords = ["smart fan", "energy efficient fan", "BLDC ceiling fan"]

all_data = []
for kw in keywords:
    google_data = scrape_google(kw)
    youtube_data = scrape_youtube_videos(kw)
    amazon_data = scrape_amazon_products(kw)
    all_data.extend(process_platform_data(google_data, "google",brands))
    all_data.extend(process_platform_data(youtube_data, "youtube",brands))
    all_data.extend(process_platform_data(amazon_data, "amazon",brands))
```

```python
from collections import defaultdict

def compute_sov_metrics(all_mentions, brands):
    counts = defaultdict(int)
    weighted_counts = defaultdict(float)
    positive_counts = defaultdict(int)

    for mention in all_mentions:
        brand = mention['brand']
        counts[brand] += 1
        weighted_counts[brand] += mention['weight']
        if mention['sentiment'] == "positive":
            positive_counts[brand] += 1

    total_mentions = sum(counts.values())
    total_weight = sum(weighted_counts.values())
    total_positive = sum(positive_counts.values())

    metrics = {}
    for brand in brands:
        metrics[brand] = {
            "Raw SoV": round((counts[brand] / total_mentions) * 100, 2) if total_mentions else 0,
            "Weighted SoV": round((weighted_counts[brand] / total_weight) * 100, 2) if total_weight else 0,
            "SoPV": round((positive_counts[brand] / total_positive) * 100, 2) if total_positive else 0
        }
    return metrics
```
